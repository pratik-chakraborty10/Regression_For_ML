# Polynomial Regression

## Importing Libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing Data Set

```
data_set=pd.read_csv("  RBC_human.csv")
X=data_set.iloc[:,0:-1].values
y=data_set.iloc[:,-1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
linear_reg=LinearRegression()
linear_reg.fit(X,y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
polynomial_reg=PolynomialFeatures(degree=6)

X_poly=polynomial_reg.fit_transform(X)

linear_reg_2=LinearRegression()
linear_reg_2.fit(X_poly,y)
```

## Visualising the Linear Regression results

```
plt.scatter(X,y,color="red")
plt.plot(X,linear_reg.predict(X),color="blue")
plt.title("RBC WITH LINEAR REGRESSION")
plt.xlabel("Age")
plt.ylabel("RBC's")
plt.show()

#The result is not perfect here.
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
plt.scatter(X,y,color="red")
plt.plot(X,linear_reg_2.predict(X_poly),color="blue")
plt.title("RBC WITH POLYNOMIAL REGRESSION")
plt.xlabel("Age")
plt.ylabel("RBC's")
plt.show()

#The result is much more perfect here.
```

## Predicting a new result with Linear Regression

```
linear_reg.predict([[6.5]]) #very wrong answer
```

## Predicting a new result with Polynomial Regression


```
linear_reg_2.predict(polynomial_reg.fit_transform([[6.5]])) #far way better result differ from linear regression
```
